```python
# pickles
# Train and Test datasets stored as pytthon pickles
import pickle
fX_train = open('fX_train.p','rb')
X_train = pickle.load(fX_train)
fX_train.close()
fy_train = open('fy_train.p', 'rb')
y_train = pickle.load(fy_train)
fy_train.close()
fX_test = open('fX_test.p', 'rb')
X_test = pickle.load(fX_test)
fX_test.close()
fy_test = open('fy_test.p','rb')
y_test = pickle.load(fy_test)
fy_test.close()
```

```python
import tensorflow as tf
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(128, activation='sigmoid', input_dim=608))
for i in range(200):
  model.add(tf.keras.layers.Dense(16, activation='sigmoid'))
  #model.add(Dropout(rate=0.1))

  # the target vector consist of 32 fields containing either a 0 or a 1
  model.add(tf.keras.layers.Dense(32, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['binary_accuracy'])
print(model.summary())
model.fit(X_train, y_train, epochs=20, batch_size=100)
model.evaluate(X_test, y_test)
predicted = model.predict(X_test)
print(predicted)
```
